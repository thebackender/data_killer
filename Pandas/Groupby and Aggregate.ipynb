```python
#task 2
account = pd.read_csv('accountancy.csv')
acc_res = account.groupby(['Executor', 'Type']).mean()
acc_res.loc[('Loopa', 'A'), 'Salary']
```

```python
arr = dict()
for i in sorted(list(account['Type'].unique())):
    if acc_res.loc[('Loopa', i), 'Salary'] > acc_res.loc[('Pupa', i), 'Salary']:
        arr[i] = 'Loopa'
    else:
        arr[i] = 'Pupa'
arr
```
